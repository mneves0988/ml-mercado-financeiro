```python
import os
import sys

# Força o PySpark a utilizar o mesmo executável Python da venv para o Driver e os Workers
os.environ["PYSPARK_PYTHON"] = sys.executable
os.environ["PYSPARK_DRIVER_PYTHON"] = sys.executable

# Resto das importações originais
from dotenv import load_dotenv
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
from pyspark.ml.feature import Tokenizer, HashingTF, IDF
from pyspark.sql.types import DoubleType
import numpy as np

# Carrega as variáveis de ambiente do arquivo .env
load_dotenv(dotenv_path='../.env')
```

```python
# DEFINIÇÃO DO JAVA 17 COMPATÍVEL (Ajuste para o seu caminho real do JDK 17 se necessário)
os.environ['JAVA_HOME'] = 'C:/Program Files/Eclipse Adoptium/jdk-17.0.19.10-hotspot'
os.environ['PATH'] = os.environ['JAVA_HOME'] + '/bin;' + os.environ['PATH']

# Variáveis do Hadoop configuradas localmente para o Windows
os.environ['HADOOP_HOME'] = 'C:/hadoop'
os.environ['PATH'] = 'C:/hadoop/bin;' + os.environ['PATH']
```

```python
# Inicializa a sessão do Spark com o driver JDBC do PostgreSQL
spark = (
    SparkSession.builder
    .appName('ML_Retrieval_TFIDF')
    .config('spark.jars.packages', 'org.postgresql:postgresql:42.7.3')
    .getOrCreate()
)

print('Sessão Spark inicializada com sucesso na versão:', spark.version)
```

```python
db_host = os.getenv('DB_HOST')
db_port = os.getenv('DB_PORT', '5432')
db_name = os.getenv('DB_NAME', 'postgres')
db_user = os.getenv('DB_USER')
db_password = os.getenv('DB_PASSWORD')

# URL de conexão JDBC para o PostgreSQL do Supabase
jdbc_url = f'jdbc:postgresql://{db_host}:{db_port}/{db_name}?sslmode=require'
```

```python
# Carrega a tabela de campanhas_links do Supabase
df_campanhas = (
    spark.read
    .format('jdbc')
    .option('url', jdbc_url)
    .option('dbtable', 'public.campanhas_links')
    .option('user', db_user)
    .option('password', db_password)
    .option('driver', 'org.postgresql.Driver')
    .load()
)

print(f"Total de campanhas carregadas para o motor de busca: {df_campanhas.count()}")
```

```python
# Mescla os campos textuais (product_type, title, description) para criar um documento unificado
df_docs = df_campanhas.withColumn(
    'document',
    F.trim(
        F.concat_ws(' ',
            F.coalesce(F.col('product_type'), F.lit('')), 
            F.coalesce(F.col('title'), F.lit('')), 
            F.coalesce(F.col('description'), F.lit(''))
        )
    )
)

# Filtra para garantir que apenas registros com algum texto valido sejam mantidos
df_docs = df_docs.filter(F.col('document') != '')
df_docs.select('product_type', 'title', 'document').show(3, truncate=False)
```

```python
# Configura e aplica o pipeline de vetorizacao TF-IDF
tokenizer = Tokenizer(inputCol='document', outputCol='tokens')
hashingTF = HashingTF(inputCol='tokens', outputCol='tf', numFeatures=20000)
idf = IDF(inputCol='tf', outputCol='tfidf')

# Transforma os textos em tokens
df_tok = tokenizer.transform(df_docs)

# Ajusta o modelo de frequencia inversa do documento (IDF)
idf_model = idf.fit(hashingTF.transform(df_tok))

# Gera os vetores TF-IDF finais para cada documento cadastrado
df_tfidf = idf_model.transform(hashingTF.transform(df_tok))
df_tfidf.cache() # Cacheia o dataframe para acelerar as multiplas consultas

print("Vetores TF-IDF gerados e armazenados em cache.")
```

```python
def _l2norm(vec):
    # Auxiliar para calcular a norma L2 do vetor (SparseVector do PySpark)
    if vec is None:
        return 0.0
    arr = vec.toArray()
    return float(np.sqrt((arr * arr).sum()))

def buscar_campanhas(query: str, top_k: int = 3):
    # 1. Converte a frase de busca do usuario em DataFrame
    q_df = spark.createDataFrame([(query,)], ['query'])
    q_df = q_df.withColumn('document', F.col('query'))
    
    # 2. Aplica as mesmas transformacoes de Tokenizer, TF e IDF
    q_tok = tokenizer.transform(q_df)
    q_tf = hashingTF.transform(q_tok)
    q_tfidf = idf_model.transform(q_tf)
    
    # Extrai o vetor gerado para a busca do usuario
    q_vec = q_tfidf.select('tfidf').first()['tfidf']
    q_norm = _l2norm(q_vec)
    
    if q_norm == 0:
        print("Busca invalida ou termos nao encontrados no vocabulario.")
        return df_docs.select('product_type', 'title', 'description').limit(top_k)
    
    # Broadcast dos vetores para as tasks executarem em paralelo de forma otimizada
    q_vec_b = spark.sparkContext.broadcast(q_vec.toArray())
    q_norm_b = spark.sparkContext.broadcast(q_norm)
    
    # UDF para calcular a similaridade de cosseno em Spark
    def cosine_similarity(v):
        if v is None:
            return 0.0
        v_arr = v.toArray()
        v_norm = float(np.sqrt((v_arr * v_arr).sum()))
        if v_norm == 0:
            return 0.0
        dot_product = float((v_arr * q_vec_b.value).sum())
        return dot_product / (v_norm * q_norm_b.value)
        
    cosine_udf = F.udf(cosine_similarity, DoubleType())
    
    # Calcula o score de similaridade para cada documento e ordena pelo maior score
    scored_df = df_tfidf.withColumn('score', cosine_udf(F.col('tfidf')))
    
    return (
        scored_df
        .select('product_type', 'title', 'description', 'score')
        .orderBy(F.col('score').desc())
        .limit(top_k)
    )
```

```python
print("--- CONSULTA DE TESTE 1 ---")
query_1 = "renda fixa conservadora"
print(f"Busca: '{query_1}'")
buscar_campanhas(query_1, top_k=3).show(truncate=False)
```

```python
print("--- CONSULTA DE TESTE 2 ---")
query_2 = "ações dividendos multimercado"
print(f"Busca: '{query_2}'")
buscar_campanhas(query_2, top_k=3).show(truncate=False)
```

```python
# Encerra a sessao do Spark
spark.stop()
print("Sessao do Spark encerrada com sucesso.")
```
